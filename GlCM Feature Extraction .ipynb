```python
glcm = feature.graycomatrix(gray_image, distances=[5], angles=[0], levels=256, symmetric=False, normed=True)
```

# Extract features from GLCM

```python
contrast = feature.graycoprops(glcm, 'contrast')
dissimilarity = feature.graycoprops(glcm, 'dissimilarity')
homogeneity = feature.graycoprops(glcm, 'homogeneity')
ASM = feature.graycoprops(glcm, 'ASM')
energy = feature.graycoprops(glcm, 'energy')
```

```python
import cv2
import numpy as np

# Load the image
img = cv2.imread('test data 1.jpeg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect keypoints and compute descriptors using the SIFT feature detector
sift = cv2.xfeatures2d.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(gray, None)

# Number of clusters
num_clusters = 100

# KMeans clustering
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
compactness, labels, centers = cv2.kmeans(descriptors, num_clusters, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# Build the histogram of features
histogram = np.zeros(num_clusters)
for i in range(len(labels)):
    histogram[labels[i]] += 1

# Normalize the histogram
histogram /= np.sum(histogram)

# The resulting histogram is the bag of visual words
bag_of_visual_words = histogram
```
